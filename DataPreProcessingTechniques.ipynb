```python
def get_parameters(df):
  parameters = {}
  for col in df.columns[df.isnull().any()]:
    if(df[col].dtype == "float64" or df[col].dtype == "int64" or df[col].dtype == "int32"):
      strategy = "mean"
    else:
      strategy = "most_frequent"
    missing_val = df[col][df[col].isnull()].values[0]
    parameters[col] = {"missing_value":missing_val , "strategy":strategy}
  return parameters
get_parameters(df)
```

```python
Parameters = get_parameters(df)
```

```python
for col,params in Parameters.items():
  missing_val = params["missing_value"]
  strategy = params["strategy"]
  imp = SimpleImputer(missing_values=missing_val , strategy=strategy)
  df[col] = imp.fit_transform(df[[col]]).ravel()
```
